```python
with open('./chat.txt', 'r', encoding='utf-8') as file:
  lines = file.readlines()
```

```python
import re
from collections import Counter

def clean_chat_data(chat_data):
    cleaned_data = []
    for line in chat_data:
        # Use regex to match lines with messages (contain ':' after name)
        if re.search(r'\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}\u202f?[ap]m - [^:]+: ', line):
            # Remove timestamp, name, and other unwanted parts
            cleaned_line = re.sub(r'\d{2}/\d{2}/\d{2}, \d{1,2}:\d{2}\u202f?[ap]m - [^:]+: ', '', line)
            # Remove media tags, emojis, and @ mentions and urls
            cleaned_line = re.sub(r'https?://\S+|www\.\S+', '', cleaned_line)
            cleaned_line = re.sub(r'<Media omitted>', '', cleaned_line)
            cleaned_line = re.sub(r'@[\w\d]+', '', cleaned_line)
            cleaned_line = re.sub(r'[^\x00-\x7f]+', '', cleaned_line)
            if(len(cleaned_line.strip())) > 0:
              cleaned_data.append(cleaned_line.strip())

    line_counts = Counter(cleaned_data)
    unique_lines = [line for line in cleaned_data if line_counts[line] <= 4]
    return unique_lines
```

```python
cleaned_lines = clean_chat_data(lines)
len(cleaned_lines)
```

```python
data = []
for l in cleaned_lines:
  data.append(l + "\n")

data[:30]
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input
```

```python
cleaned_data = [line.strip().lower() for line in data]
full_text = " ".join(cleaned_data)
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([full_text])
```

```python
(tokenizer.word_index)
```

```python
total_words = len(tokenizer.word_index) + 1
total_words
```

```python
# convert into supervised learning task
input_sequences = []
for line in cleaned_data:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)
```

```python
input_sequences[:10]
```

```python
max_seq_len = max(len(x) for x in input_sequences)
max_seq_len
```

```python
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre'))
```

```python
print(input_sequences.shape)
input_sequences[:10]
```

```python
# Split into training X and y and testing
```

```python
X = input_sequences[:,:-1]
y = input_sequences[:,-1]
y = to_categorical(y, num_classes=total_words)
```

```python
print(X.shape, y.shape)
```

```python
model = Sequential()
model.add(Embedding(total_words, 100))  # Input shape is (None, 108)
model.add(LSTM(150, return_sequences=True))  # First LSTM returns sequences
model.add(LSTM(150))  # Second LSTM
model.add(Dense(total_words, activation='softmax'))  # Output layer
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X, y, batch_size=256, epochs=20, verbose=1)
```
